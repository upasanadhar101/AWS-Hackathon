```python
#GROUPING BY Country and cities and taking mean of individual pollutants in individual cities 

df_new_mean_2015 = df_2015.groupby(['Country', 'City', 'Specie'])[ "max"].mean()
df_new_mean_2016 = df_2016.groupby(['Country', 'City', 'Specie'])[ "max"].mean()
df_new_mean_2017 = df_2017.groupby(['Country', 'City', 'Specie'])[ "max"].mean()
df_new_mean_2018 = df_2018.groupby(['Country', 'City', 'Specie'])[ "max"].mean()
df_new_mean_2019 = df_2020.groupby(['Country', 'City', 'Specie'])[ "max"].mean()
df_new_mean_2020 = df_2020.groupby(['Country', 'City', 'Specie'])[ "max"].mean()
df_new_mean_2015.head
```

```python
#GROUPING BY Country and cities and taking mean of all the pollutants (cumultaive average of pollutants ) in that individual cities 

df_new_2_mean_2015 = df_2015.groupby(['Country', 'City'])["Specie", "max"].mean()
df_new_2_mean_2016 = df_2016.groupby(['Country', 'City'])["Specie", "max"].mean()
df_new_2_mean_2017 = df_2017.groupby(['Country', 'City'])["Specie", "max"].mean()
df_new_2_mean_2018 = df_2018.groupby(['Country', 'City'])["Specie", "max"].mean()
df_new_2_mean_2019 = df_2019.groupby(['Country', 'City'])["Specie", "max"].mean()
df_new_2_mean_2020 = df_2020.groupby(['Country', 'City'])["Specie", "max"].mean()

df_new_2_mean_2015.head()
```
